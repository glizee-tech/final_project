```python
import requests

files = [
    {"name": "incidents.csv", "url": "http://groupe1-achat.onrender.com/public/incidents.csv"},
    {"name": "orders.json", "url": "http://groupe1-achat.onrender.com/public/orders.json"},
    {"name": "suppliers.parquet", "url": "http://groupe1-achat.onrender.com/public/suppliers.parquet"},
]

base_path = "/Volumes/workspace/final_project/landing/"

for f in files:
    r = requests.get(f["url"], stream=True, timeout=60)
    r.raise_for_status()

    with open(base_path + f["name"], "wb") as out:
        for chunk in r.iter_content(chunk_size=1024 * 1024):
            if chunk:
                out.write(chunk)

print("Landing completed")
```
